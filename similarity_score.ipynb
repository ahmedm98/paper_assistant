```python
import pandas as pd
from pathlib import Path
import numpy as np
from numpy.linalg import norm
```

```python
papers_dataset_path=Path("data/updated_papers_dataset.csv")
papers_dataset = pd.read_csv(papers_dataset_path)
papers_dataset=papers_dataset.astype('object')
```

```python
papers_dataset
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
papers_embeddings=[]
for i in papers_dataset.index:

    #retrieve extracted paper
    paper_summary = papers_dataset.iloc[i]["summary"]

    embedding = model.encode(paper_summary)
    papers_embeddings.append(list(embedding))
```

```python
papers_dataset["summary_embedding"]=papers_embeddings
```

```python
papers_dataset
```

```python
csv_path=Path("data/updated_papers_dataset2.csv")
papers_dataset.to_csv(csv_path,index=False)
```
