# Preprocess the extracted text

```python
import pandas as pd
from pathlib import Path 
```

```python
papers_dataset_path=Path("data/papers_dataset.csv")
papers_dataset = pd.read_csv(papers_dataset_path)
extracted_texts_path=Path("data/extracted-text")
preprocessed_texts_path=Path("data/preprocessed-text")
```

```python
for i in papers_dataset.index:

    #retrieve extracted paper
    paper_name = papers_dataset.iloc[i]["paper"]

    with open(Path(extracted_texts_path,paper_name+'.txt')) as f:
        paper_content = f.read()
        f.close()

    # Preprocess the extracted text to remove the abstracts, title lines, tables and empty lines.
    sectioned_content=paper_content.split("----")

    if sectioned_content[0].split(" ")[0].rstrip() == "ABSTRACT":
        papers_dataset.at[i,"abstract"] = sectioned_content[0].replace("ABSTRACT","").replace("\n","").strip()
    
    #save the preprocessed texts in a separate folder
    with open(Path(preprocessed_texts_path,paper_name+'.txt'),"a") as f:

        for s in sectioned_content:
                       
            #remove tables
            if s.strip().startswith("DESCRIPTION TABLE:") or s.strip().startswith("ABSTRACT"):
                pass
            else:
                lines=s.strip().splitlines(True)
                f.write("".join([l.replace("\n","") for l in lines[1:]]))
        f.close()

    
```
